```python
import matplotlib.pyplot as plt
import time
import sys
def fibonacci(n):
  a,b=0,1
  seq = []
  for _ in range(n):
    seq.append(a)
    a,b=b,a+b
  return seq
```

```python
def merge_sort(arr):
    if len(arr) > 1:
        mid = len(arr) // 2
        L = arr[:mid]
        R = arr[mid:]
        merge_sort(L)
        merge_sort(R)
        i = j = k = 0
        while i < len(L) and j < len(R):
            if L[i] < R[j]:
                arr[k] = L[i]
                i += 1
            else:
                arr[k] = R[j]
                j += 1
            k += 1
        while i < len(L):
            arr[k] = L[i]
            i += 1
            k += 1
        while j < len(R):
            arr[k] = R[j]
            j += 1
            k += 1
    return arr
```

```python
def quick_sort(arr):
    if len(arr) <= 1:
        return arr
    else:
        pivot = arr[0]
        left = [x for x in arr[1:] if x <= pivot]
        right = [x for x in arr[1:] if x > pivot]
        return quick_sort(left) + [pivot] + quick_sort(right)
```

```python
def insertion_sort(arr):
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1
        while j >= 0 and key < arr[j]:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key
    return arr
```

```python
def binary_search(arr, target):
    left, right = 0, len(arr) - 1
    while left <= right:
        mid = (left + right) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            left = mid + 1
        else:
            right = mid - 1
    return -1
#testing
sizes = [10, 50, 100, 200, 500]
time_results = {"merge_sort": [], "quick_sort": [], "insertion_sort": [], "fibonacci": []}
for n in sizes:
    test_arr = list(range(n, 0, -1))
```

```python
# testing
sizes = [10, 50, 100, 200, 500]
time_results = {"merge_sort": [], "quick_sort": [], "insertion_sort": [], "fibonacci": []}
for n in sizes:
    test_arr = list(range(n, 0, -1))
    # Fibonacci
    start = time.time()
    fibonacci(n)
    end = time.time()
    time_results["fibonacci"].append(end - start)
    # Merge Sort
    start = time.time()
    merge_sort(test_arr.copy())
    end = time.time()
    time_results["merge_sort"].append(end - start)
    # Quick Sort
    start = time.time()
    quick_sort(test_arr.copy())
    end = time.time()
    time_results["quick_sort"].append(end - start)
    # Insertion Sort
    start = time.time()
    insertion_sort(test_arr.copy())
    end = time.time()
    time_results["insertion_sort"].append(end - start)

# plot using matplot lib
plt.figure(figsize=(10,6))
for algo, times in time_results.items():
    plt.plot(sizes, times, marker='o', label=algo)
plt.title("Algorithm Time Complexity (Empirical)")
plt.xlabel("Input Size (n)")
plt.ylabel("Time (seconds)")
plt.legend()
plt.grid(True)
plt.show()
# viualizing
algos = ["Fibonacci", "Merge Sort", "Quick Sort", "Insertion Sort", "Binary Search"]
space = ["O(1)", "O(n)", "O(log n)", "O(1)", "O(1)"]
plt.figure(figsize=(8,5))
plt.bar(algos, [1, 2, 1.5, 1, 1], color='skyblue')
plt.title("Space Complexity (Relative)")
plt.ylabel("Relative Memory Usage")
plt.show()
```
